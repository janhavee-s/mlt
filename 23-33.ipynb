```python
# Task 23: k-NN Accuracy

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

df = pd.read_csv("insurance.csv")

df['target'] = (df['charges'] > df['charges'].median()).astype(int)

le = LabelEncoder()
for col in ['sex', 'smoker', 'region']:
    df[col] = le.fit_transform(df[col])

X = df.drop(['charges', 'target'], axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Task 24: k-NN k=1 to 10

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

df = pd.read_csv("insurance.csv")

df['target'] = (df['charges'] > df['charges'].median()).astype(int)

le = LabelEncoder()
for col in ['sex', 'smoker', 'region']:
    df[col] = le.fit_transform(df[col])

X = df.drop(['charges', 'target'], axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

for k in range(1, 11):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    print(f"k={k}, Accuracy={accuracy_score(y_test, pred)}")
```

```python
# Task 25: k-NN Scaling Effect

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

df = pd.read_csv("insurance.csv")

df['target'] = (df['charges'] > df['charges'].median()).astype(int)

le = LabelEncoder()
for col in ['sex', 'smoker', 'region']:
    df[col] = le.fit_transform(df[col])

X = df.drop(['charges', 'target'], axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Without scaling
model1 = KNeighborsClassifier(n_neighbors=3)
model1.fit(X_train, y_train)
pred1 = model1.predict(X_test)

# With scaling
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

model2 = KNeighborsClassifier(n_neighbors=3)
model2.fit(X_train_s, y_train)
pred2 = model2.predict(X_test_s)

print("Without scaling:", accuracy_score(y_test, pred1))
print("With scaling:", accuracy_score(y_test, pred2))
```

```python
# Task 30: Linear vs Polynomial Regression

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df = pd.read_csv("insurance.csv")

le = LabelEncoder()
for col in ['sex', 'smoker', 'region']:
    df[col] = le.fit_transform(df[col])

X = df.drop('charges', axis=1)
y = df['charges']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Linear
lin = LinearRegression()
lin.fit(X_train, y_train)
pred1 = lin.predict(X_test)

# Polynomial
poly = PolynomialFeatures(degree=2)
X_train_p = poly.fit_transform(X_train)
X_test_p = poly.transform(X_test)

lin2 = LinearRegression()
lin2.fit(X_train_p, y_train)
pred2 = lin2.predict(X_test_p)

print("Linear MSE:", mean_squared_error(y_test, pred1))
print("Polynomial MSE:", mean_squared_error(y_test, pred2))
```

```python
# Task 31: Regression Curve (Fixed)

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression

df = pd.read_csv("insurance.csv")

le = LabelEncoder()
for col in ['sex', 'smoker', 'region']:
    df[col] = le.fit_transform(df[col])

X = df[['age']]
y = df['charges']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)

# Sort values for proper line
X_test_sorted = X_test.sort_values(by='age')
y_pred_sorted = model.predict(X_test_sorted)

# Plot
plt.scatter(X_test, y_test)

plt.plot(X_test_sorted, y_pred_sorted)

plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Regression Curve")

plt.show()
```
